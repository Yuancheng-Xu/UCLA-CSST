```r
library(REEMtree)
source("SimData.r")
```

```r
# simulate the data
n = 100
T = 5
set.seed(100)
data = sim_3(n,T)
data = data.frame(data)

# features name are V1,...,V400 
names(data) = c(paste("V",1:400,sep=""),"y")

# add patient/cluster information
for (i in 1:n){
    data$ID[(1+(i-1)*T):(i*T)] = rep(i,T)
}

# see the data
head(data)
```

```r
# we use all the features here
variables = paste("V",1:400,sep="")
# formula
Formular = as.formula(paste("y~",paste(variables,collapse = "+")))
# Formular
```

```r
system.time({REEMresult<-REEMtree(Formular, data=data, random=~1|ID)})
plot(REEMresult)
```

```r
predict(REEMresult,data)
```

```r
# The first way to get important features, which does not match the plot of tree
sort(REEMresult$Tree$variable.importance, decreasing = TRUE)
```

```r
# Another way: extract the features used in the node:
(REEMresult$Tree$frame$var)
```
